```python
# Import the necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
sns.set(color_codes=True)
```

```python
# Load the car.csv dataset
car = pd.read_csv('car.csv')
```

```python
# Display the head of the dataset
car.head()
```

```python
# Dsplay the bottom 5 rows
car.tail(5) 
```

```python
# Check the data types of all the columns
car.dtypes
```

```python
# Display all the columns
car.columns
```

```python
# Analytical summary of the dataset
car.describe()
```

```python
car.hist(figsize=(20,30));
```

```python
# Relationship between categorical and continuous variable
sns.boxplot(x="Vehicle Size", y="Engine HP", data = car);
```

```python
sns.pairplot(car)
```

```python
# Drop irrelevant columns
car = car.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
car.head(5)
```

```python
# Renaming the column names
car = car.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
car.head(5)
```

```python
# Total number of rows and columns
car.shape
```

```python
# Rows containing duplicate data
duplicate_rows_car = car[car.duplicated()]
print("Number of duplicate rows: ", duplicate_rows_car.shape)
```

```python
# Count the rows before removing the data
car.count()
```

```python
# Drop the duplicates 
car = car.drop_duplicates()
car.head()
```

```python
# Count the number of rows after removing duplicates.
car.count()
```

```python
# Find the null values
car.isna().sum()
```

```python
# Find the null values
print(car.isnull().sum())
```

```python
# Drop the missing values.
car = car.dropna() 
car.count()
```

```python
# Count after dropping the values
print(car.isnull().sum()) 
```

```python
# Finding the outliers
sns.boxplot(x=car['Price']);
```

```python
sns.boxplot(x=car['HP']);
```

```python
# Plotting a Histogram for number of cars per brand
car.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel("Number of cars")
plt.xlabel("Make");
```

```python
# Find the relations between the variables.
plt.figure(figsize=(20,10))
c= car.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
